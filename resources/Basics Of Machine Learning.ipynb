## Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[10,20,25,30,40]
```

```python
plt.plot(x,y)
plt.title('Line plot')
plt.xlabel('x -axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python
months=['Jan','Feb','March','April','May','June']
revenue=[1000,2000,3000,4000,5000,6000]
plt.bar(months,revenue)
plt.title('Sales Chart')
plt.xlabel('Months')
plt.ylabel('Revenue in Crores')
plt.show();
```

```python
marks50=np.random.randint(0,100,50)
marks50
```

```python
plt.hist(marks50,bins=5)
plt.title('Marks distribution')
plt.xlabel('Marks Range')
plt.ylabel('Count')
plt.show();
```

```python
pheight=[65,70,68,72,66]
pweight=[150,160,155,100,145]
```

```python
plt.scatter(pheight,pweight)
plt.title('Height v/s Weight')
plt.xlabel('Height in Inches')
plt.ylabel('Weight in pounds')
plt.show();
```

```python
days=[1,2,3,4,5]
original_price=[10,20,25,30,40]
predicted_price=[8,21,25,28,38]
```

```python
plt.figure(figsize=(10,6))
plt.plot(days,original_price,'b-o',label='Original Price')
plt.plot(days,predicted_price,'r-o',label='Predicted Price')
plt.legend()
plt.title('Stock Price Prediction')
plt.xlabel('days')
plt.ylabel('Price')
plt.show();
```

```python
plt.savefig('Stock_Prediction_Chart.png',dpi=300)
```
